```python
import logging
logging.basicConfig(
    filename="char_frequency.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
def character_frequency(s: str) -> dict:
    """
    Count the frequency of each character in a given string.
    Parameters:
        s (str): Input string
    Returns:
        dict: Dictionary containing character frequencies.
    """
    logging.info("Calculating character frequency")
    freq = {}
    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    logging.info(f"Character frequency calculated: {freq}")
    return freq
s = input("Enter string: ")
result = character_frequency(s)
print("Character frequency:")
for char, count in result.items():
    print(f"{char} : {count}")
```
