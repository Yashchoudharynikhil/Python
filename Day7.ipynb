***JASON FILE***

```python
data={
    "name":"yash",
    "age": 22,
    "course":"CSE"
}
```

```python
import json
with open('json1.json','w') as f:
    json.dump(data,f)
```

```python
with open('json1.json','r') as f:
    #json.dump(data,f)
    print(json.load(f))
```

```python
with open('json1.json','w+') as f:
    json.dump(data,f)
    f.seek(0)
    print(json.load(f))
```

```python
with open('json1.json','r+') as f:
    json.dump(data,f)
    f.seek(0)
    da=json.load(f)
    print(da)
    print(len(da))
```

***CSV***

```python
import csv
with open("data,csv","w",newline="") as file:
    writer=csv.writer(file)
    writer.writerow(["name","age","course"])
    writer.writerow(["yash",22,"cse"])
```

```python
row=[
    ["name","age"],
    ["yash","22"],
    ["ravi","33"]
]
with open("data.csv","w",newline="") as file:
    writer= csv.writer(file)
    writer.writerow(row)
```

```python
import csv
with open("data1.csv","w",newline="") as file:
    fieldnames=["name","age"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"name":"yash","age":22})
    writer.writerow({"name":'kk',"age":22})
```

```python
with open("data1.csv","r")as file:
    reader=csv.DictReader(file)
    print(reader)
    for row in reader:
        print(row)
```
