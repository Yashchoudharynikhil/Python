```python
#create a list of square of even numbers only from a given list
def sq(l)->list:
       return[x**2 for x in l]
arr=[1,2,3,4]
print(sq(arr))
```

```python
#extract all words starting with vowel from a sentance
def extract(s)->list:
    vowel="aeiouAEIOU"
    return[x for x in s.split() if x[0] in vowel]
print(extract("hello this is yash kumar"))
```

```python
# Remove duplicate elements from a list using a set and print only value greater than 5
lst=[1,3,3,5,7,2]
remove=[num for num in set(lst) if num>5]
print(remove)
```
