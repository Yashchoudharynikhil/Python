```python
#question 3 of function
def palindrome(num1):
    n1 = num1
    rev = 0
    while num1>0:
        ldigit = num1%10
        rev = rev*10 + ldigit
        num1 = num1//10
    return "Palindrome" if rev==n1 else "Not Palindrome"
print(palindrome(1211))
```
