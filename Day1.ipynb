```python
#Given a string: 
# If the first character is uppercase, print "Starts with uppercase" 
# Else if the last character is lowercase, print "Ends with lowercase" 
# Else print "Other case"

a=input("Enter the string")
if a[0].isupper():
    print("Starts with upperCase")
elif a[-1].islower():
    print("Ends with lower")
else:
    print("Other case")
```

```python
#Given a list of integers: 
#• If the sum of first half > sum of second half, print "First half greater" 
#• Else if equal, print "Equal halves" 
#• Else print "Second half greater" 
lst = [1, 1, 1, 2]
mid = len(lst) // 2
first_sum = sum(lst[:mid])
second_sum = sum(lst[mid:])
if first_sum > second_sum:
    print("First half greater")
elif first_sum == second_sum:
    print("Equal halves")
else:
    print("Second half greater")
```

```python
#Given a string: 
# If it starts with "A" 
# If it ends with "Z" → print "Valid AZ string" 
# Else → print "Starts with A but invalid end" 
# Else → print "Invalid string"
str=input("Enter input")
if str[0]=='A':
    if str[-1]=='Z':
        print("Valid AZ string")
    else:
        print("Starts with A but invalid end")
else:
    print("Invalid string")
```

```python
#Given a string: 
# If the first 2 characters are digits 
# Check whether the last 2 characters are alphabets 
# Else print "Invalid format"
s = input("Enter input: ")

if s[:2].isdigit():
    if s[-2:].isalpha():
        print("Valid format")
    else:
        print("Last 2 characters are not digits")
else:
    print("invalid format")
```

```python
#Given a list: 
# If the list length is even 
# Compare first 2 elements and last 2 elements, Slice with same element otherwise slice with 
# different elements. 
# Else print "Odd length list" 
nums = [5, 6, 1, 2]
if len(nums) % 2 == 0: 
    if nums[:2] == nums[-2:]: 
        print("Slices are the same")
    else:
        print("Slices are different")
else:
    print("Odd length list")
```

```python
#Given a list: 
#If the list length is even 
# Compare first 2 elements and last 2 elements, Slice with same element otherwise slice with 
# different elements. 
# Else print "Odd length list"
def count_vowels(s) -> int:
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

string = input("Enter a string: ")
count = count_vowels(string)
print("Number of vowels in the string:", count)
```
