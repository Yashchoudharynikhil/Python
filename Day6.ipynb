```python
#public Access specifier
class A:
    x=10
    def display(self):
        print(self.x)
obj=A()
print(obj.x)
obj.display()
        
```

```python
#protected Access specifier

class B(A):
    def show(self):
        print(self._x)
obj =B()
obj.show()
```

```python
#private Access specifier
class A:
    def __init__(self):
        self.__x =30
    def show(self):
        print(self.__x)
obj =A()
obj.show()
```

```python
from abc import ABC , abstractmethod

class BankAccount(ABC):
    def __init__(self,customer_name,date_of_birth):
        self.customer_name = customer_name
        self.date_of_birth = date_of_birth

    @abstractmethod
    def deposit(self,amount):
        pass

    @abstractmethod
    def withdraw(self,amount):
        pass

class SavingAccount(BankAccount):
    def __init__(self,customer_name,date_of_birth,minimum_balance,balance):
        super().__init__(customer_name,date_of_birth)
        self.minimum_balance=minimum_balance
        self.balance=balance
        

    def deposit(self,amount):
        if(amount>0):
            self.balance+=amount
            print(f"{amount} deposit sucessfull")
        else:
            print("amount should be positive")
    def withdraw(self,amount):
        if (amount>self.balance-self.minimum_balance):
            print("amount should be less")
        else:
            self.balance-=amount
            print(f"remaining amount {self.balance}")




```

```python
obj1=SavingAccount("abc","01-12-",1000,0)
obj1.withdraw(1000)
obj1.deposit(100000)
```

```python
class CurrentAccount(BankAccount):
    def __init__(self,customer_name,date_of_birth,overdraft,balance):
        super().__init__(customer_name,date_of_birth)
        self.overdraft=overdraft
        self.balance=balance
        

    def deposit(self,amount):
        if(amount>0):
            self.balance+=amount
            print(f"{amount} deposit sucessfull")
        else:
            print("amount should be positive")
    def withdraw(self,amount):
        if (amount>self.balance-self.minimum_balance):
            print("amount should be less")
        else:
            self.balance-=amount
            print(f"remaining amount {self.balance}")
```
